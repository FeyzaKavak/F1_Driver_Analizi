```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline
```

```python
dataFrame = pd.read_csv('F1Drivers_Dataset.csv')
dataFrame.head()
```

```python
dataFrame = dataFrame.dropna(axis=1) # NaN degerleri attik
```

```python
dataFrame.describe()
```

```python
NationalityDF = dataFrame['Nationality'] # Yarismaciların ulkelerini aldik.
UlkeListesi=set(NationalityDF)
UlkeListesi

```

```python
len(UlkeListesi) #47 tane ulke kayitli
```

```python
ChampionshipsDF = dataFrame['Championships']
NCDataFrame = pd.concat([NationalityDF,ChampionshipsDF], axis=1)
NCDataFrame
```

```python
groupbyDF = dataFrame.groupby(['Nationality']).Championships.agg([len,sum]) 
groupbyDF
# ulkelerin kac defa yarısmaya katildiği gösteriliyor (len)
# ulkelerin kac toplam kac defa sampiyon oldugunu gosteriyor (sum)
```

```python
plt.figure(figsize=(10,15))
NCDataFrame["Nationality"].value_counts().plot.barh().set_title("Ulkelerin yarismalara katilim sayisi")
```

```python
SampiyonDF = dataFrame[dataFrame['Champion'] == True]
SampiyonDF
# sadece sampiyon olanlari aldik
```

```python
SampiyonDF.describe()
#34 sampiyon var
```

```python
set(SampiyonDF['Nationality'])
# Sampiyonlugu olan ulkeler
```

```python
Nationality_verisi = SampiyonDF['Nationality']
deger_sayisi = Nationality_verisi.value_counts()
plt.figure(figsize=(10, 10))  # Grafigin boyutu
plt.pie(deger_sayisi, labels=deger_sayisi.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # daire olarak cizmek için
plt.show()
# Sampiyonlar icinde En cok hangi ulkeden sampiyon ciktigini gosteriyor
```

```python
Championships = dataFrame['Championships']
Race_Wins = dataFrame['Race_Wins']
korelasyonRC = Race_Wins.corr(Championships)
print('Sampiyonluk ile kazanilan yarislarin arasinda ki korelasyon iliskisi:\n',korelasyonRC)

# aralarında ki iliski yuksek diyebiliriz
```

```python
Race_Entries = dataFrame['Race_Entries']
Race_Wins = dataFrame['Race_Wins']
korelasyonRR = Race_Wins.corr(Race_Entries)
print('Yaris girisleri ile kazanilan yarislarin arasinda ki korelasyon iliskisi:\n',korelasyonRR)

# aralarinda iliski var diyebiliriz.
```

```python
Championships = dataFrame['Championships']
Race_Entries = dataFrame['Race_Entries']
korelasyonCR = Race_Entries.corr(Championships)
print('Sampiyonluk ile yarisa girislerin arasinda ki korelasyon iliskisi:\n',korelasyonCR)

# aralarinda iliski var ancak cok yuksek degil
```

```python
Fastest_Laps = dataFrame['Fastest_Laps']
Race_Wins= dataFrame['Race_Wins']
korelasyonFR = Race_Wins.corr(Fastest_Laps)
print('Hizli tur sayisi ile kazanilan yaris arasinda ki korelasyon iliskisi:\n',korelasyonFR)

# aralarinda ki iliski yuksek diyebiliriz
```

```python
Fastest_Laps = dataFrame['Fastest_Laps']
Championships= dataFrame['Championships']
korelasyonFC = Championships.corr(Fastest_Laps)
print('Hizli tur sayisi ile sampiyonluk arasinda ki korelasyon iliskisi:\n',korelasyonFC)

# aralarinda ki iliski yuksek sayilir
```

```python
C = dataFrame['Championships']
RW = dataFrame['Race_Wins']
RE = dataFrame['Race_Entries']
FL = dataFrame['Fastest_Laps']

dataFrameCRRF = pd.concat([C,RW,RE,FL],axis=1)
korelasyon_matrisi = dataFrameCRRF.corr()
```

```python
plt.figure(figsize=(8, 6))
sbn.heatmap(korelasyon_matrisi, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelasyon Haritasi\n')
plt.show()
```
